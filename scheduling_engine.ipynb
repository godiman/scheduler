```python
import pandas as pd
import numpy as np
import random
```

```python
data = pd.DataFrame({'doctor':['d1', 'd2', 'd3', 'd4', 'd5'], 
                     'p1':['2021-04-01 00:15:30', '2021-04-01 00:45:02', '2021-04-01 00:30:22',
                           '2021-04-01 00:52:04', '2021-04-01 00:47:17'],
                     'p2':['2021-04-01 00:22:45', '2021-04-01 00:35:01', '2021-04-01 00:17:48',
                           '2021-04-01 00:49:20', '2021-04-01 00:47:53'],
                     'p3':['2021-04-01 00:52:02', '2021-04-01 00:29:42', '2021-04-01 00:43:10',
                           '2021-04-01 00:24:19', '2021-04-01 00:19:53'], 
                     'p4':['2021-04-01 00:33:24', '2021-04-01 00:48:02', '2021-04-01 00:29:10',
                           '2021-04-01 00:41:20', '2021-04-01 00:33:48']
                     })
```

```python
data
```

```python

data['p1'] = pd.to_datetime(data['p1'])
data['p2'] = pd.to_datetime(data['p2'])
data['p3'] = pd.to_datetime(data['p3'])
data['p4'] = pd.to_datetime(data['p4'])
```

```python
data.describe()
```

```python
data['p1_minute'] = data['p1'].dt.minute
data['p2_minute'] = data['p2'].dt.minute
data['p3_minute'] = data['p3'].dt.minute
data['p4_minute'] = data['p4'].dt.minute

# print(minute)
data
```

```python
data['avg_in_min'] = data.mean(axis=1).plot()
# print(avg)
```

```python
data
```

```python
data['total_time_min']  = data[['p1_minute', 'p2_minute', 'p3_minute', 'p4_minute']].sum(axis=1)
# print(total_time)
data
```

```python
# Maximum number of a patient a doctor can take
data['max_p'] = data['total_time_min'].div(data['avg_in_min'])
# print(p_max)
data
```
